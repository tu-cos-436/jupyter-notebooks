## Software Stack for Data Analysis and Visualization in Python

These are some of the Python packages and other software that we will be using.

1. [NumPy](https://numpy.org/) ("Numerical Python") 
   brings fast and efficient
   multi-dimensional arrays to Python.
1. [pandas](https://pandas.pydata.org/) (originally from "panel data")
   provides data analysis and manipulation,
   simplifying manipulation of tabular and similar structured data;
   relies heavily on NumPy.
1. [matplotlib](https://matplotlib.org/) ("Mathematical Plotting Library")
   is a package for (mostly) two-dimensional plotting
1. [seaborn](https://seaborn.pydata.org/)
   provides higher-level graphing capability for "statistica data visualization;"
   it runs atop `matplotlib`
1. [IPython](https://ipython.org/) ("Interative Python")
   aims to be a "rich architecture for interactive computing."
   Started as a "better Python," but now aims to be a language-agnostic platform.
   Python is still the best-supported language.
1. [Jupyter](https://jupyter.org/)
   is a platform for multi-language interactive computing;
   supports Python and R and a bevy of other languages
   
These are just some of the very common packages and platforms.

# NumPy

First, we import the Python Numeric package, `numpy`. Convention dictates importing it as `np`.

```python
import numpy as np
```

## NumPy is fast

```python
py_list = list(range(1_000_000))  # Python list
np_array = np.arange(1_000_000)   # NumPy list
```

IPython supports "magic" commands, which start with a `%`.
For example, `%time` measures the run time of the command it prefixes.

We use this to measure how long it takes to double the elements in each list.

Start with the NumPy array. Note that we can operate on `numpy` arrays with built-in operators
like `*` for multiplication.

```python
%time for _ in range(50): np_array = np_array * 2
```

Now repeat the operation on a native Python list.

```python
%time for _ in range(50): py_list2 = [x * 2 for x in py_list]
```

## NumPy handles matrices

```python
data = np.random.randn(8, 4)
data
```

```python
data * 10
```

## NumPy matrices are typed

```python
np.array([1, 3, 5, 3, 7, 9], dtype=np.float64)
```

```python
np.array([1, 3, 5, 3, 7, 9], dtype=np.int32)
```

## NumPy does arithmetic

```python
data = np.array([[1, 3, 5], [4, 6, 8]])
data
```

```python
data * data
```

```python
data + (data * 2)
```

```python
data.T @ data
```

## NumPy operates on entire arrays

```python
points = np.arange(-5, 5, 0.5)
points
```

```python
xs, ys = np.meshgrid(points, points)
xs
```

```python
ys
```

```python
z = np.sqrt(xs ** 2 + ys ** 2)
z
```

## NumPy can drive visualizations

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(z, cmap=plt.cm.Blues)
plt.colorbar()
plt.title("Plot of $\sqrt{x^2 + y^2}$ for grid of values")
```

## 1D pandas object is a `Series`

```python
obj = pd.Series([1, 2, 3, 5, 8, 13])
obj
```

A `Series` object has an array of values and an index (here a numeric range)

```python
obj.values
```

The `index` need not be numeric. Here, construct a `Series` from a Python dictionary.

## pandas loads data

Take a look at the first couple of lines of a CSV containing GapMinder data.

```python
!head -2 gapminder_all.csv
```

```python
gm = pd.read_csv('gapminder_all.csv')
gm
```

## pandas filters columns (similar to SQL `SELECT`)

```python
gm[['continent', 'country']]
```

## pandas filters rows (similar to SQL `WHERE`)

```python
africa_boolean = gm['continent'] == 'Africa'
```

```python
africa = gm[africa_boolean][['country', 'pop_2007']]
africa
```

## pandas will sort

```python
africa_sorted = africa.sort_values(by="pop_2007", ascending=False)
africa_sorted
```

## pandas supports visualization

```python
africa_sorted.plot(kind='barh', legend=False)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12, 11))
sns.barplot(data=africa_sorted, x='pop_2007', y='country')
```
