```python
locations = site.drop_duplicates().head(1000)
locations
```

```python
for loc in locations.itertuples():
    print(loc.LATITUDE, loc.LONGITUDE)
```

# Geospatial

```python
import folium
```

```python
map = folium.Map(location=[35.02262, -93.47239])
for loc in locations.itertuples():
    folium.Marker([loc.LATITUDE, loc.LONGITUDE]).add_to(map)
map
```
